# Load the data about the farmers

Data about the loans and about the ward locations of the farmers were provided separately. Load those datasets and join them on the farmer index.

```python
farmer_loans = pd.read_csv(os.path.join(DATA_ROOT,
                           'regression',
                           'farmer_loan_results.csv'))

farmer_loans.set_index('fmr_id', inplace=True)

print(farmer_loans.shape)
farmer_loans.head()
```

```python
farmer_wards = pd.read_csv(os.path.join(DATA_ROOT,
                                        'regression',
                                        'farmer_wards.csv'),
                           index_col=0)

farmer_wards.set_index('fmr_id', inplace=True)
farmer_wards.drop('loanid', axis=1, inplace=True)
farmer_wards.dropna(inplace=True)

farmer_wards = farmer_wards.reset_index().drop_duplicates().set_index('fmr_id')

print(farmer_wards.shape)
farmer_wards.head()
```

```python
all_farmer_data = farmer_loans.join(farmer_wards, how='left')

all_farmer_data.columns.tolist()

all_farmer_data.shape
```

```python
all_farmer_data.head()
```

# Load the satellite data

Load data from the satellite imagery:

 - `planet`: normalized indices calculated from RapidEye imagery for Fall of 2016
 - `modis`: indices from MODIS data for March of 2017


```python
planet = pd.read_csv(os.path.join(DATA_ROOT,
                                  'regression',
                                  'all_ward_data_55wards.csv'))

planet.columns = ['clean_ward'] + planet.columns[1:].tolist()
planet.head()
```

```python
modis = pd.read_csv(os.path.join(DATA_ROOT,
                                 'regression',
                                 'march17_ward_data_modis.csv'),
                     index_col=0)
modis.head()
```

```python
modis['clean_ward'] = (modis.index
                      .to_series()
                      .str
                      .split('.')
                      .apply(
            lambda x: x[1].replace('/', '-').replace(' ', '-')))

# modis.set_index('clean_ward', inplace=True)
```

## Check ward overlap

Make sure that the wards have been updated so that they match across datasets.

```python
all_farmer_data['clean_ward'] = (all_farmer_data.ward
                                                .apply(
              lambda x: str(x).replace('/', '-').replace(' ', '-')))
```

```python
# planet data is missing 6 of the wards we see in the farmer data
np.setdiff1d(all_farmer_data.clean_ward,
             planet.clean_ward)
```

```python
# modis has all of the wards that we see in the farmer data22
np.setdiff1d(all_farmer_data.clean_ward,
             modis.clean_ward)
```

# Features

## Build Loan features

Drop features that:
 - only have a single value
 - are just unique identifiers rather than data about the loan
 - drop features that we wouldn't have ahead of time (i.e., when we decide to make the loan)
 
Only look at non-rollover loans, since (presumably those are the ones we care about).

```python
# number of unique values in each column per loan
all_farmer_data.groupby('loan_id').nunique().nunique()
```

```python
farmer_features = all_farmer_data.reset_index().groupby('loan_id').agg({
    'loanstatus_id': 'first',  # these are all the same, see check above
    'rollstatus': 'first',
    'loan_date': 'first',
    'loan_amnt_requested': 'first',
    'loan_issued': 'first',
    'loan_principal': 'first',
    'loan_interest': 'first',
    'loan_repaystart': 'first',
    'loan_created': 'first',
    'clean_ward': 'first'
})
```

```python
print(farmer_features.shape)
farmer_features = farmer_features[farmer_features.clean_ward != 'nan']
print(farmer_features.shape)
```

```python
# 3 is repaid, 4 is abandoned
target_loan_repayment = (farmer_features.loanstatus_id == 3.0).astype(int)

farmer_features = farmer_features.drop('loanstatus_id', axis=1)
```

```python
farmer_features.dtypes
```

# Modeling

```python
from sklearn.preprocessing import StandardScaler

# drop co-linear vars
X = farmer_features.drop(['loan_interest', 'loan_principal', 'loan_repaystart', 'clean_ward'],
                         axis=1)

ss = StandardScaler()
X_vals = ss.fit_transform(X)


X_vals = pd.DataFrame(data=X_vals, columns=X.columns, index=X.index)
```

## Model Just Loan Features

```python
import statsmodels.formula.api as sm

model = sm.Logit(target_loan_repayment, X_vals)

result = model.fit_regularized(full_output=True)
result.summary()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate


scores = cross_validate(LogisticRegression(), X_vals, target_loan_repayment, scoring='roc_auc',
                         cv=10, return_train_score=True)

np.mean(scores['test_score'])
```

## Model Loan + MODIS Features

```python
# drop co-linear vars
X = farmer_features.drop(['loan_interest', 'loan_principal', 'loan_repaystart'],
                         axis=1)

print(X.shape)
X = X.merge(modis, on='clean_ward')
print(X.shape)

X.drop('clean_ward', inplace=True, axis=1)

ss = StandardScaler()
X_vals = ss.fit_transform(X)


X_vals = pd.DataFrame(data=X_vals, columns=X.columns, index=target_loan_repayment.index)
```

```python
X_vals.head()
```

```python
model = sm.Logit(target_loan_repayment, X_vals)

result = model.fit(method='bfgs', full_output=True)
result.summary()
```

```python
scores = cross_validate(LogisticRegression(), X_vals, target_loan_repayment, scoring='roc_auc',
                         cv=10, return_train_score=True)

np.mean(scores['test_score'])
```

## Model Loan + planet Features

```python
# drop co-linear vars
X = farmer_features.drop(['loan_interest', 'loan_principal', 'loan_repaystart'],
                         axis=1)


print(X.shape)
X = X.merge(planet, how='left', on='clean_ward')
print(X.shape)

X.fillna(-10, inplace=True)

X.drop('clean_ward', inplace=True, axis=1)

ss = StandardScaler()
X_vals = ss.fit_transform(X)

X_vals = pd.DataFrame(data=X_vals, columns=X.columns, index=target_loan_repayment.index)
```

```python
model = sm.Logit(target_loan_repayment, X_vals)

result = model.fit(method='bfgs', full_output=True)
result.summary()
```

```python
scores = cross_validate(LogisticRegression(), X_vals, target_loan_repayment, scoring='roc_auc',
                         cv=10, return_train_score=True)

np.mean(scores['test_score'])
```

## Model Loan + planet + MODIS Features

```python
# drop co-linear vars
X = farmer_features.drop(['loan_interest', 'loan_principal', 'loan_repaystart'],
                         axis=1)


print(X.shape)
X = X.merge(planet, how='left', on='clean_ward')
X = X.merge(modis, on='clean_ward')
print(X.shape)

X.fillna(-10, inplace=True)

X.drop('clean_ward', inplace=True, axis=1)

ss = StandardScaler()
X_vals = ss.fit_transform(X)

X_vals = pd.DataFrame(data=X_vals, columns=X.columns, index=target_loan_repayment.index)
```

```python
model = sm.Logit(target_loan_repayment, X_vals)

result = model.fit(method='newton', full_output=True)
result.summary()
```

```python
scores = cross_validate(LogisticRegression(), X_vals, target_loan_repayment, scoring='roc_auc',
                         cv=10, return_train_score=True)

np.mean(scores['test_score'])
```
